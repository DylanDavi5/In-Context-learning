```python
import requests
from io import BytesIO
from zipfile import ZipFile
import os

# URL of the file to download
url = "https://github.com/dtsip/in-context-learning/releases/download/initial/models.zip"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract the content of the response
    zip_content = BytesIO(response.content)

    # Open the Zip file
    with ZipFile(zip_content, 'r') as zip_ref:
        # Specify the directory where you want to extract the contents
        print(f"Current working directory: {os.getcwd()}")
        !ls
        print("Extracted files:", os.listdir("gdrive/MyDrive/Project_deep_neural_networks/in-context-learning/src/models"))
        extraction_path = "gdrive/MyDrive/Project_deep_neural_networks/in-context-learning/src"

        # Extract all the contents of the zip file to the specified directory
        zip_ref.extractall(extraction_path)

    print("Extraction completed.")
else:
    # Print an error message if the request was not successful
    print(f"Failed to download the file. Status code: {response.status_code}")
```
