```python
def merge_dicts(dict1, dict2):
    merged_dict = dict1.copy()

    for key, value in dict2.items():
        if key in merged_dict and isinstance(merged_dict[key], dict) and isinstance(value, dict):
            # Recursive merge for nested dictionaries
            merged_dict[key] = merge_dicts(merged_dict[key], value)
        elif key in merged_dict and isinstance(merged_dict[key], list) and isinstance(value, list):
            # Extend lists if both values are lists
            merged_dict[key].extend(value)
        else:
            # Non-dictionary values or new key
            merged_dict[key] = value

    return merged_dict
```

```python
location="gdrive/MyDrive/Project_deep_neural_networks/in-context-learning/src/" #change this to relevant file structure

parser = argparse.ArgumentParser()

parser.add_argument("--config", type=str, required=True)

args = parser.parse_args(["--config", location+"conf/decision_tree_and_linear.yaml"])

with open(args.config, "r") as file:
    config_data = yaml.safe_load(file)
    if "inherit" in config_data:
        for file2 in config_data["inherit"]:
            file2mark=location+"conf/"+file2
            with open(file2mark, "r") as file2_content:
                content_data = yaml.safe_load(file2_content)
                config_data=merge_dicts(config_data, content_data)

    config_data["out_dir"]=location+config_data["out_dir"][3:]
    print(config_data)
    config = MainConfig(**config_data)

assert config.model.family in ["gpt2", "lstm"]
print(f"Running with: {config}")

if not config.test_run:
    run_id = config.training.resume_id or str(uuid.uuid4())
    out_dir = os.path.join(config.out_dir, run_id)
    if not os.path.exists(out_dir):
        os.makedirs(out_dir)
    config.out_dir = out_dir

    with open(os.path.join(out_dir, "config.yaml"), "w") as yaml_file:
        yaml.dump(config.dict(), yaml_file, default_flow_style=False)

main(config)
```
